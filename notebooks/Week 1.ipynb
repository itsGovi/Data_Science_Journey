#### Define the Objective

You want to extract meaningful insights from numerical columns by grouping, analyzing distributions, and checking correlations. Your goal is to uncover patterns such as:

- Salary differences across departments

- Utilization trends

- Performance score impact on flight risk

---

Columns used to do above things:

`base_salary`, `billing_rate`, `performance_score`, `actual_utilization`, `utilization_target`, `flight_risk`

```python
"""
By Department

- What's the average salary per department?
- Which department has the highest billing rate?
- Does utilization vary significantly across departments?
"""

# 1️⃣ Average salary per region
avg_salary_by_region = df.groupby('region')['base_salary'].mean()


# 2️⃣ Region with highest billing rate
higest_billing_rate = df.groupby('region')['billing_rate'].idxmax()


# 3️⃣ utilization vary significantly across departments
util_variation_stats = df.groupby('department')[['utilization_target','actual_utilization']].agg(['mean', 'std'])
round(util_variation_stats, 2)
```

```python
"Q: Do senior employees have significantly higher base salaries?"

avg_salary_by_level = df.groupby('level')['base_salary'].mean()
avg_salary_by_level.median
```

```python
sns.boxplot(x='level', y='base_salary', data=df)

plt.title("Base Salary Distribution by Level")
plt.xlabel("Employee Level")
plt.ylabel("Base salary")

plt.show()
```

```python
"Q: Are higher-level employees more utilized than juniors?"
filtered_df = df[df['level'].isin(['entry', 'senior'])]
avg_util_by_levels = filtered_df.groupby(['level', 'department'])['actual_utilization'].agg(['mean', 'median'])
avg_util_by_levels

# [OR]

"""
avg_util_by_levels = (
    df[df['level'].isin(['entry', 'senior'])]
    .groupby(['level', 'department'])['actual_utilization']
    .agg(['mean', 'median'])
)
"""
```

```python
"Q: Do high performers have better salaries?"
def compare_salaries(df):
    # overall salary stats
    overall_stats = df.groupby(['level', 'department'])['base_salary'].agg(['mean', 'median'])

    # high performance emp's
    high_perf_df = df[df['performance_score'] >= 4]

    # high performance emp's stats
    high_perf_stats = high_perf_df.groupby(['level', 'department'])['base_salary'].agg(['mean', 'median']).reset_index()

    # merge the two results for side by side comparision
    comparision_df = overall_stats.merge(high_perf_stats, on=['level', 'department'], suffixes=('_overall', '_high'))

    correlation = comparision_df['mean_overall'].corr( comparision_df['mean_high'] )

    # returning the merged df
    return comparision_df, correlation

result_df, corr_value = compare_salaries(df)
```
